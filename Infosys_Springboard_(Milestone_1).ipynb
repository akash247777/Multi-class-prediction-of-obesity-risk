```python
# Importing the pandas library for data manipulation and analysis.
import pandas as pd

# Importing the numpy library for numerical operations.
import numpy as np

# Importing the matplotlib library for creating visualizations.
import matplotlib.pyplot as plt
```

```python
path='/content/drive/MyDrive/Infosys obesity dataset_raw.csv'
```

```python
df=pd.read_csv(path)
```

```python
df.head()
```

```python
df=df.rename(columns={'FAVC' : 'frequent_consumption_of_high_caloric_food',
'FCVC' : 'frequency_of_consumption_of_vegetables',
'NCP' : 'number_of_main_meals',
'CAEC' : 'consumption_of_food_between_meals',
'CH20' : 'consumption_of_water_daily',
'CALC' : 'consumption_of_alcohol',
'SCC' : 'calories_consumption_monitoring',
'FAF' : ' physical_activity_frequency',
'TUE' : 'time_using_technology_devices',
'MTRANS' : 'transportation_used',
'NObeyesdad' : 'obesity_level'})
```

```python
df.head()
```

```python
df.shape
```

```python
# Finding duplicates
duplicates = df[df.duplicated()]

print("Duplicate rows:")
print(duplicates)
```

```python
# Removing duplicates
df_cleaned = df.drop_duplicates()

print("\nDataFrame after removing duplicates:")
print(df_cleaned)
```

```python
df_cleaned.shape
```

```python
df_cleaned[-5:]
```

```python
# Checking for missing values (NaN) in each column of the cleaned DataFrame.
df_cleaned.isna().sum()
```

```python
# Selecting rows from df_cleaned that have missing values (NaN) or contain spaces (' ') in any column.
df_cleaned[df_cleaned.isna().any(axis=1) | (df_cleaned == ' ').any(axis=1)]
```

```python
# Filling all NaN values in df_cleaned with np.nan
df_cleaned = df_cleaned.fillna(np.nan)

# Filling NaN values in the 'frequency_of_consumption_of_vegetables' column with the mean of the column
df_cleaned['frequency_of_consumption_of_vegetables'] = df_cleaned['frequency_of_consumption_of_vegetables'].fillna(df_cleaned['frequency_of_consumption_of_vegetables'].mean())
```

```python
df_cleaned.iloc[12:16,:]
```

```python
# Selecting rows from df_cleaned that have missing values (NaN) or contain spaces (' ') in any column.
df_cleaned[df_cleaned.isna().any(axis=1) | (df_cleaned == ' ').any(axis=1)]
```

```python
# Replacing NaN values in 'family_history_with_overweight' column with the mode (most frequent value)
df_cleaned['family_history_with_overweight'] = df_cleaned['family_history_with_overweight'].fillna(df_cleaned['family_history_with_overweight'].mode()[0])
```

```python
# Selecting rows from df_cleaned that have missing values (NaN) or contain spaces (' ') in any column.
df_cleaned[df_cleaned.isna().any(axis=1) | (df_cleaned == ' ').any(axis=1)]
```

```python
df_cleaned.iloc[12:16,:]
```

```python
# Initializing an empty list to store unique values of each feature
unique_values_list = []

# Defining the list of columns to drop from the DataFrame
drop_columns = ['id', 'Age', 'Height', 'Weight', 'frequency_of_consumption_of_vegetables',
                'number_of_main_meals', 'CH2O', 'physical_activity_frequency',
                'time_using_technology_devices']

# Creating a list of features by excluding the columns in drop_columns
features = [col for col in df_cleaned.columns if col not in drop_columns]

# Iterating over each feature and appending its unique values to unique_values_list
for column in features:
    unique_values_list.append({column: df_cleaned[column].unique().tolist()})

# Outputting the list of unique values for each feature
unique_values_list
```

```python
# Iterating over each feature and printing the count of unique values in each column
for col in features:
    print(df_cleaned[col].value_counts(), '\n\n')
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a 3x3 subplot grid
fig, axes = plt.subplots(3, 3, figsize=(25, 15))
fig.subplots_adjust(hspace=0.5, wspace=0.5)  # Adjust spacing between subplots

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through each feature
for i, feature in enumerate(features):
    ax = axes[i]  # Select the current axis

    # Get unique values of the current feature
    unique_values = df_cleaned[feature].unique()

    # Plot bar plot for the current feature
    ax.bar(unique_values, df_cleaned[feature].value_counts(), color=['red','green','yellow','blue','pink','orange','grey'], alpha=0.7)
    ax.set_xlabel(feature)
    ax.set_ylabel('Count')
    ax.set_title('Bar Plot - ' + feature)
    ax.grid(True)

# Remove empty subplots
for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
